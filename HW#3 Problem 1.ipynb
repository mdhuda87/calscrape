```python
import requests
req = requests.get("https://www.ccny.cuny.edu/registrar/fall-2018-academic-calendar")
```

```python
from bs4 import BeautifulSoup as BS
import pandas as pd
import numpy as np
```

```python
soup = BS(req.text, 'html.parser')
soup
```

```python
table = soup.select('tr')
len(table)
```

```python
record = [record for record in table]
data = record[1:]
data
```

```python
values = [[record.get_text().strip() for record in record.find_all('td')] for record in data]
values[30]
```

```python
df = pd.DataFrame(values, columns = ['dates','dow', 'text'])
df.replace('', np.nan, inplace=True)
df.dropna(inplace=True)
df.set_index('dates')
```
